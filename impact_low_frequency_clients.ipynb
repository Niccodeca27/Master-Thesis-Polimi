```python
import pickle

# Load the pickles
with open('cluster_dicts/cluster_dict_k-means_stretched_one_centroid_inversely_proportional_v6(k=50).pkl', 'rb') as f:
    dict_data = pickle.load(f)
with open('cluster_dicts/cluster_dict_k-means_stretched_one_centroid_inversely_proportional_v6(k=50)_AS.pkl', 'rb') as f:
    dict_data_AS = pickle.load(f)
with open('cluster_dicts/cluster_dict_k-means_stretched_one_centroid_inversely_proportional_v6(k=50)_ON.pkl', 'rb') as f:
    dict_data_ON = pickle.load(f)

    
with open('client_groups_full.pkl', 'rb') as f:
    client_groups = pickle.load(f)

# Filter out elements from dict values that are in client_groups
filtered_dict = [[loc_id for loc_id in lista if loc_id not in client_groups['0 volte']] for lista in dict_data.values() if len(lista)>0]
filtered_dict_AS = [[loc_id for loc_id in lista if loc_id not in client_groups['0 volte']] for lista in dict_data_AS.values() if len(lista)>0]
filtered_dict_ON = [[loc_id for loc_id in lista if loc_id not in client_groups['0 volte']] for lista in dict_data_ON.values() if len(lista)>0]

filtered_dict = [elem for elem in filtered_dict if len(elem)>0]
filtered_dict_ON = [elem for elem in filtered_dict_ON if len(elem)>0]
filtered_dict_AS = [elem for elem in filtered_dict_AS if len(elem)>0]
```

```python
pc.calc_clusters_stats(filtered_dict).to_csv('clustering_methods_performances/k-means_stretched_one_centroid_inversely_proportional_v6(k=50)_LFIMPACT.csv')
pc.calc_clusters_stats_AS(filtered_dict_AS).to_csv('clustering_methods_performances/k-means_stretched_one_centroid_inversely_proportional_v6(k=50)_AS_LFIMPACT.csv')
pc.calc_clusters_stats_ON(filtered_dict_ON).to_csv('clustering_methods_performances/k-means_stretched_one_centroid_inversely_proportional_v6(k=50)_ON_LFIMPACT.csv')
```
